# Plotting Extracted Features from Images of Flowers

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
from mpl_toolkits.mplot3d import Axes3D
```

```python
roses = pd.read_csv('rose_features.csv')
chrysanthemums = pd.read_csv('chrysanthemum_features.csv')
butterfly_pea = pd.read_csv('butterflypea_features.csv')
```

## eccentricity vs r

```python
plt.figure(figsize=(8,6))

plt.scatter(roses['r'],roses['eccentricity'],c='red',s=30,label='roses')
plt.scatter(chrysanthemums['r'],chrysanthemums['eccentricity'],c='orange',s=30,label='chrysanthemums')
plt.scatter(butterfly_pea['r'],butterfly_pea['eccentricity'],c='purple',s=30,label='butterfly peas')
plt.ylabel('eccentricity',fontsize=14)
plt.xlabel('r',fontsize=14)
plt.legend(fontsize=14)

# plt.savefig('evsr.png',dpi=300,bbox_inches='tight')
plt.show()
```

## eccentricity vs g

```python
plt.figure(figsize=(8,6))

plt.scatter(roses['g'],roses['eccentricity'],c='red',s=30,label='roses')
plt.scatter(chrysanthemums['g'],chrysanthemums['eccentricity'],c='orange',s=30,label='chrysanthemums')
plt.scatter(butterfly_pea['g'],butterfly_pea['eccentricity'],c='purple',s=30,label='butterfly peas')
plt.ylabel('eccentricity',fontsize=14)
plt.xlabel('g',fontsize=14)
plt.legend(fontsize=14)

# plt.savefig('evsg.png',dpi=300,bbox_inches='tight')
plt.show()
```

## eccentricity vs b

```python
plt.figure(figsize=(8,6))

plt.scatter(roses['b'],roses['eccentricity'],c='red',s=30,label='roses')
plt.scatter(chrysanthemums['b'],chrysanthemums['eccentricity'],c='orange',s=30,label='chrysanthemums')
plt.scatter(butterfly_pea['b'],butterfly_pea['eccentricity'],c='purple',s=30,label='butterfly peas')
plt.ylabel('eccentricity',fontsize=14)
plt.xlabel('b',fontsize=14)
plt.legend(fontsize=14)

# plt.savefig('evsb.png',dpi=300,bbox_inches='tight')
plt.show()
```

## r vs g vs b

```python
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(roses['r'], roses['g'], roses['b'], c='red', marker='o',label='roses',s=40)
ax.scatter(chrysanthemums['r'], chrysanthemums['g'], chrysanthemums['b'], c='orange', marker='o',label='chrysanthemums',s=40)
ax.scatter(butterfly_pea['r'], butterfly_pea['g'], butterfly_pea['b'], c='purple', marker='o',label='butterfly peas',s=40)

ax.set_xlabel('r',fontsize=16)
ax.set_ylabel('g',fontsize=16)
ax.set_zlabel('b',fontsize=16)

plt.legend(fontsize=14)
# plt.savefig('rgb.png',dpi=300,bbox_inches='tight')
plt.show()
```
