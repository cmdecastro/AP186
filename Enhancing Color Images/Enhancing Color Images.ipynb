```python
import numpy as np
import matplotlib.pyplot as plt
```

## Read and display test images

```python
img = plt.imread('trial7.jpeg')

plt.figure(figsize=(8,6))
plt.imshow(img),plt.xticks([]),plt.yticks([])
plt.show()
```

## Contrast Stretching

```python
def stretching(img):
    R = img[:,:,0]
    G = img[:,:,1]
    B = img[:,:,2]

    I_stretch = np.copy(img)
    
    R_stretch = (R-np.amin(R))/(np.amax(R)-np.amin(R))
    G_stretch = (G-np.amin(G))/(np.amax(G)-np.amin(G))
    B_stretch = (B-np.amin(B))/(np.amax(B)-np.amin(B))
    
    I_stretch[:,:,0] = 255*R_stretch
    I_stretch[:,:,1] = 255*G_stretch
    I_stretch[:,:,2] = 255*B_stretch
    
    return np.round(I_stretch)
```

```python
plt.figure(figsize=(8,6))

plt.imshow(stretching(img)),plt.xticks([]),plt.yticks([])
# plt.savefig('14_cs.png',dpi=300,bbox_inches='tight')

plt.show()
```

## Grayworld Algorithm

```python
def grayworld(img):
    R = img[:,:,0]
    G = img[:,:,1]
    B = img[:,:,2]
    
    R_ave = np.mean(R)
    G_ave = np.mean(G)
    B_ave = np.mean(B)
    
    R_wb = R/R_ave
    G_wb = G/G_ave
    B_wb = B/B_ave
    
    avg = np.mean(img)
#     avg = np.max([R_ave,G_ave,B_ave])
    
    I_wb = np.copy(img)

    I_wb[:,:,0] = avg*R_wb
    I_wb[:,:,1] = avg*G_wb
    I_wb[:,:,2] = avg*B_wb
    
    return np.round(I_wb)
```

```python
plt.figure(figsize=(8,6))

plt.imshow(grayworld(img)),plt.xticks([]),plt.yticks([])
# plt.savefig('14_gw.png',dpi=300,bbox_inches='tight')

plt.show()
```
